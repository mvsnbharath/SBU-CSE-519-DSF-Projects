```python
## Custom Functions to explore data of dtype into, float and onject (categorical values)

def exploreIntColums(columnName,columnData,totalEntries):
    missingEntries = totalEntries - columnData.count()
    tempDataObject = {}
    tempDataObject['Column Name'] = columnName
    tempDataObject['dType'] = 'int'
    tempDataObject['Missing Values(Count)'] = missingEntries
    tempDataObject['% of Missing Values'] = float(str(round(missingEntries*100/totalEntries,2)))
    tempDataObject['Mean'] = columnData.mean()
    tempDataObject['Standard Deviation'] = columnData.std()
    return tempDataObject

def exploreFloatColumns(columnName,columnData,totalEntries):
    missingEntries = totalEntries - columnData.count()
    tempDataObject = {}
    tempDataObject['Column Name'] = columnName
    tempDataObject['dType'] = 'float'
    tempDataObject['Missing Values(Count)'] = missingEntries
    tempDataObject['% of Missing Values'] = float(str(round(missingEntries*100/totalEntries,2)))
    tempDataObject['Mean'] = columnData.mean()
    tempDataObject['Standard Deviation'] = columnData.std()
    return tempDataObject

def convertCountToPercentage(dictOfFrequencyCount,totalEntries):
    return {k: float(str(round(v*100/ totalEntries, 2)))  for k, v in dictOfFrequencyCount.items()}

def findNMostFrequentValuesAndCount(columnData,n,totalEntries):
    
    if(n>3):
        frequencyCountDict = columnData.value_counts().nlargest(n=3).to_dict()
        return convertCountToPercentage(frequencyCountDict,totalEntries)
    else:
        frequencyCountDict = columnData.value_counts().nlargest(n).to_dict()
        return convertCountToPercentage(frequencyCountDict,totalEntries)
    
def exploreObjectColumns(columnName,columnData,totalEntries):
    missingEntries = totalEntries - columnData.count()
    tempDataObject = {}
    tempDataObject['Column Name'] = columnName
    tempDataObject['dType'] = 'string'
    tempDataObject['Unique Values'] = columnData.nunique()
    tempDataObject['Most Frequent Values'] = findNMostFrequentValuesAndCount(columnData,columnData.nunique(),totalEntries)
    tempDataObject['Missing Values(Count)'] = missingEntries
    tempDataObject['% of Missing Values'] = float(str(round(missingEntries*100/totalEntries,2)))
    return tempDataObject
```

```python
def exploreData(train):
    totalEntries = train.shape[0]
    
    dataInt = pd.DataFrame([])
    dataFloat = pd.DataFrame([])
    dataObj = pd.DataFrame([])
    for x in train.columns:
        if train[x].dtype.name == 'int64':
            dataInt = dataInt.append(exploreIntColums(x,train[x],totalEntries),ignore_index=True )
        elif train[x].dtype.name == 'float64':
            dataFloat = dataFloat.append(exploreFloatColumns(x,train[x],totalEntries),ignore_index=True )
        elif train[x].dtype.name == 'object':
            dataObj = dataObj.append(exploreObjectColumns(x,train[x],totalEntries),ignore_index=True )
        else:
            print("other type")
    return dataInt,dataFloat,dataObj
```

### Get Highest and Least correlations

* From the above matrix, we can get the highest and least correlated values
* Optimization:
    * Removing diaognal elements ad corr(a,a)=1
    * Remove lower triangle as, corr(a,b)= corr(b,a)

```python
def getRedundantPairs(df):
    '''
    Removing diaognal elements and lower trianglular values
    As, corr(a,b)= corr(b,a) and corr(a,a)=1
    '''
    pairsToDrop = set()
    cols = df.columns
    for i in range(0, df.shape[1]):
        for j in range(0, i+1):
            pairsToDrop.add((cols[i], cols[j]))
    return pairsToDrop

def getHighestAndLowestCorrelations(df, n=5):
    corr = df.corr().unstack()
    labelsToDrop = getRedundantPairs(df)
    corr = corr.drop(labels=labelsToDrop).sort_values(ascending=True)
    positiveCorr = corr[corr>0].sort_values(ascending=False)
    negativeCorr = corr[corr<0].sort_values(ascending=True)
    return positiveCorr[0:n], negativeCorr[0:n]
```

```python
positiveCorr, negativeCorr = getHighestAndLowestCorrelations(customTrainDf,3)
display("Most Positive Correlations (we have taken top 3)")
display(positiveCorr)
display("Most Negative Correlations (we have taken top 3)")
display(negativeCorr)
```

## Part 3 - Handcrafted Scoring Function

### Dealing with Missing Values

* Null values for this data set are not simply missing values. I inferred this from the data description.
* Let me explain this with an example:
    * Some columns that which appear as null might legitimately be known (ie: "na").
    * Example: **GarageYrBlt has 5.55% na values, we have to infer it as there are houses with no Garage rather than seeing it as data is incomplete**.
    * There's a difference between NULL (unknown), and "NA" (not available). It might mean the difference between "there isn't a garage" and "there is no garage data".
    * <b> We will consider both test and train data here, so that we can define how to deal with missing values in those columns.<br>
    (The idea behind this is, there could be few columns which may not have na values in training set but you may find them in test set </b>
    
* Fill null values for each column by using domain knowledge (mean,median, mode, '0',etc)

# We will define what to do in each row by reading the data description

* **
* 12) &emsp; Electrical: NA means 'no electrical'

* **All fields related to basement, replacement  with 'no veneer'**
* 28) &emsp; MasVnrArea: None means 'no veneer' -> Replace with '0'
* 29) &emsp; MasVnrType: None means 'no veneer'
* **
* 30) &emsp; MiscFeature: NA means 'None'
* 31) &emsp; PoolQC: NA means 'no pool'
* 32) &emsp; SaleType:  -> Replace with mode
* 33) &emsp; TotalBsmtSF: NA means 'no basement'  -> Replace with '0'
* 34) &emsp; Utilities: -> Replace with mode

```python
def replace_missing_values (df):
    df['Alley']        = df['Alley'].fillna('no alley')
    df['BsmtCond']     = df['BsmtCond'].fillna('no basement')
    df['BsmtExposure'] = df['BsmtExposure'].fillna('no basement')
    df['BsmtFinSF1']   = df['BsmtFinSF1'].fillna(0)
    df['BsmtFinSF2']   = df['BsmtFinSF2'].fillna(0)
    df['BsmtFinType1'] = df['BsmtFinType1'].fillna('no basement')
    df['BsmtFinType2'] = df['BsmtFinType2'].fillna('no basement')
    df['BsmtFullBath'] = df['BsmtFullBath'].fillna(0)
    df['BsmtHalfBath'] = df['BsmtHalfBath'].fillna(0)
    df['BsmtQual']     = df['BsmtQual'].fillna('no basement')
    df['BsmtUnfSF']   =  df['BsmtUnfSF'].fillna(0)
    df['Electrical']   = df['Electrical'].fillna('no electrical')
    df['Exterior1st']  = df['Exterior1st'].fillna('no exterior')
    df['Exterior2nd']  = df['Exterior2nd'].fillna('no exterior')
    df['Fence']        = df['Fence'].fillna('no fence')
    df['FireplaceQu']  = df['FireplaceQu'].fillna('no fireplace')
    df['Functional']   = df['Functional'].fillna(df.Functional.mode())
    df['GarageArea']   = df['GarageArea'].fillna(0)
    df['GarageCars']   = df['GarageCars'].fillna(0)
    df['GarageCond']   = df['GarageCond'].fillna('no garage')
    df['GarageFinish'] = df['GarageFinish'].fillna('no garage')
    df['GarageQual']   = df['GarageQual'].fillna('no garage')
    df['GarageType']   = df['GarageType'].fillna('no garage')
    df['GarageYrBlt']  = df['GarageYrBlt'].fillna(np.nanmedian(df['GarageYrBlt']))
    df['KitchenQual']  = df['KitchenQual'].fillna(df.KitchenQual.mode())
    df['LotFrontage']  = df['LotFrontage'].fillna(0)
    df['MSZoning']     = df['MSZoning'].fillna(df.MSZoning.mode())
    df['MasVnrArea']   = df['MasVnrArea'].fillna(0)
    df['MasVnrType']   = df['MasVnrType'].fillna('None')
    df['MiscFeature']  = df['MiscFeature'].fillna('None')
    df['PoolQC']       = df['PoolQC'].fillna('no pool')
    df['SaleType']     = df['SaleType'].fillna(df.SaleType.mode())
    df['TotalBsmtSF']  = df['TotalBsmtSF'].fillna(0)
    df['Utilities']    = df['Utilities'].fillna(df.Utilities.mode())
    return df
```

## Part 4 - Pairwise Distance Function

* We will use the ranking function to determine pairwise correlations to determine the distance functions

| House No | Column A | Column B    |
|----------|----------|-------------|
|   h1      |    a1    |     b1     |
|   h2      |    a2    |     b2     |

<br>

<center> <b> a1,a2,b1,b2 are all scaled values for all the columns we are interested. </b> </center>

<center> <b> dist(h1=h2) = $ \sqrt{(a1-a2)^{2} + (b1-b2)^{2}}$  </b> </center>


```python
def preProcessData(df):
    df = df.fillna(df.median())
    df = df.fillna('None')
    le = LabelEncoder()
    df = df.apply(le.fit_transform)
    df.drop(['PoolQC'], axis=1, inplace=True)
    df.drop(['Alley'], axis=1, inplace=True)
    df.drop(['MiscFeature'], axis=1, inplace=True)
    df.drop(['Fence'], axis=1, inplace=True)
    df.drop(['Id'], axis=1, inplace=True)
    return df
```

```python
df = pd.read_csv('Data/train.csv')
df = preProcessData(df)
```

#### Encode the Values and scale them

```python
min_max_scaler = MinMaxScaler()
df_scaled = pd.DataFrame(min_max_scaler.fit_transform(df.values))
```

#### Calculate the euclidean distance

```python
Y = pdist(df_scaled, 'seuclidean', V=None)
disanceMatrix = squareform(Y)
print(disanceMatrix.shape)
```

```python
def perform_pca_tne(df):
    
    pca = PCA(n_components=50)
    pca_result = pca.fit_transform(df)
    print('Reduced number of features:', pca_result.shape)
    print("======PCA done=========")
    time_start = time.time()
    tsne = TSNE(n_components=2, verbose=1, perplexity=40, n_iter=300)
    tsne_results = tsne.fit_transform(pca_result)
    print('t-SNE task done! Time elapsed: {} seconds'.format(time.time()-time_start))
    print("=======t-SNE done=======")
    
    return tsne_results
    
```

```python
def data_pre_processing_vanilla_model(df):
    df = replace_missing_values(df)
    df = df.fillna(df.median())
    df = df.fillna('None')
    df.drop(['Id', 'SalePrice'], axis = 1,inplace=True) 
    le = LabelEncoder()
    df = df.apply(le.fit_transform)
    ss = MinMaxScaler()
    df_scaler = pd.DataFrame(ss.fit_transform(df))
    columnNames = df.columns.to_list()
    df_scaler.columns = columnNames
    return df_scaler
```

```python
def plot_top_coefficients(model, top_n = 10):
    '''
    Plots the magnitude of top and bottom n coefficients
    '''
    cols = X_train.columns
    coef = model.coef_
    zipped = list(zip(cols, coef))
    zipped.sort(key=lambda x: x[1], reverse = True)
    top_coefficients = pd.DataFrame(zipped).head(top_n).sort_values(1)
    bottom_coefficients = pd.DataFrame(zipped).tail(top_n).sort_values(1,ascending=True)
    combined = pd.concat([bottom_coefficients,top_coefficients], axis=0)
    combined.columns = ['Feature', 'Coefficient']
    return combined
```

```python
combined= plot_top_coefficients(lmPred, 5)
plt.figure(figsize=(10,6))
ax = sns.barplot(x='Coefficient', y='Feature', data=combined)
ax.set_ylabel('Features')
plt.xlabel("Linear Regressionn Coefficients Value",fontsize = 12)
plt.title("Top Linear Regression Coefficients ",fontsize = 20)
plt.show()
```

## Part 8 - Permutation Test

```python
from sklearn.model_selection import permutation_test_score
from sklearn.metrics import make_scorer
```

```python
def rmsle(true, pred):
    return -np.sqrt(metrics.mean_squared_error(np.log(true), np.log(pred)))

def data_pre_process(df):
    df = df.fillna(df.median())
    df = df.fillna('None')
    le = LabelEncoder()
    df = df.apply(le.fit_transform)
    return df

def compute_pvalue_and_plot(df,columnsOfIntrest):
    
    for column in list(columnsOfIntrest):
        variable = df.filter([column])
        X_train, X_test, y_train, y_test = train_test_split(variable,result, test_size=0.2)
        lm = LinearRegression()
        rmsle_score = make_scorer(rmsle, greater_is_better=False)
        score, permutation_scores, pvalue = permutation_test_score(lm, X_train, y_train,
                                                                   scoring = rmsle_score,cv=3, 
                                                                   n_permutations=100, n_jobs=1)

        plt.hist(permutation_scores, 20, label='Permutation scores',
             edgecolor='black')
        ylim = plt.ylim()
        plt.plot(2 * [score], ylim, '--g', linewidth=3,
             label='Classification Score'
             ' (pvalue %s)' % pvalue)
   
        plt.ylim(ylim)
        plt.legend()
        plt.xlabel('Score')
        plt.show()
```

```python
df = pd.read_csv('Data/train.csv')
result = train['SalePrice']
df = data_pre_process(df)
columnsOfIntrest=['MiscVal','YrSold','SaleType','ExterQual','KitchenQual',
                 'TotalBsmtSF','HeatingQC','BsmtQual','TotRmsAbvGrd','YearBuilt']
compute_pvalue_and_plot(df,columnsOfIntrest)
```
